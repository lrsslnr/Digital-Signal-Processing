```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fs = 40
Ts = 1/fs
f0 = 2
n = np.arange(fs)
xn = np.cos(2 * np.pi * f0 * n * Ts)
k = 1
df = k * fs
x2n = np.cos(2 * np.pi * (f0+df) * n * Ts)


fs2 = 400
Ts2 = 1/fs2
n2 = np.arange(fs2)
xfs2n = np.cos(2 * np.pi * (f0+df) * n2 * Ts2)
```

```python
# Inicia um novo plot
plt.figure()
# Cria um sinal
plt.plot(n*Ts, xn, label = "Sinal Amostrado")
plt.plot(n*Ts, x2n,'o', label = "Sinal com Aliasing")
plt.plot(n2*Ts2, xfs2n, label = "Sinal com freq alta")
#Cria legenda
plt.ylabel('Magnitude')
plt.xlabel('Tempo (s)')
plt.legend()
plt.show()
```
