```python
# Cria visualização sem Popup
%matplotlib inline
# Manipulação de gráficos
import matplotlib.pyplot as plt
# Manipulação de números
import numpy as np
```

$x[n] = x(n T_s)$

$x(t) = \cos(2 \pi f _0 t)$

$x[n] = \cos(2 \pi f_o n T_s)$

```python
# Frequência de Amostragem em Hz
fs = 44100
# Inverso da frequência de Amostragm
Ts = 1/fs
# Frequência fundamental
f0 = 440
# Range da frequência de amostragem
n = np.arange(44100)
xn = np.cos(2 * np.pi * f0 * n * Ts)
```

```python
# Inicia um novo plot
plt.figure()
# Cria um sinal
plt.plot(n[0:200]*Ts, xn[0:200], label = "Sinal Amostrado")
#Cria legenda
plt.ylabel('Magnitude')
plt.xlabel('Tempo (s)')
plt.show()
```
