```python
import matplotlib.pyplot as plt

# Inicia um novo plot
plt.figure()
# Cria um sinal
plt.plot([0, 1, 2, 3, 4, 5], [-1, 1, -1, 3, -1, 4])
#Cria legenda
plt.ylabel('Magnitude')
plt.xlabel('Medida')
plt.show()

plt.figure()
# Dois sinais com legenda
plt.plot([0, 1, 2, 3, 4, 5], [-1, 1, -1, 3, -1, 4], label = "Azul")
plt.plot([0, 1, 2, 3, 4, 5], [-2, 3, 1, -1, 2, 1], label = "Laranja")
plt.ylabel('Magnitude')
plt.xlabel('Medida')
plt.legend()
plt.show()
```
